```python
#exercice 1
import numpy as np
import matplotlib.pyplot as plt
A= np.ones((10, 10), dtype=bool)
plt.imshow(A);
```

```python
def cadre(B):
    U=B
    n=np.size(B[0,:])
    U[1,:]=np.zeros(n)
    U[-2,:]=np.zeros(n)
    U[:,1]=np.zeros(n)
    U[:,-2]=np.zeros(n)
    plt.imshow(U);

cadre(A)
```

```python
#exercice 2
random=np.random.randint(0,2,size=(6,8))
plt.imshow(random)

def cadre2(B):
    n=int(np.size(B[:,0]))
    m=int(np.size(B[0,:]))
    mat=np.zeros((n+2,m+2))
    mat[1:-1,1:-1]=B
    plt.imshow(mat)

cadre2(random)
```

```python
#exercice 3
def anti_diagonal(L):
    n=len(L)
    mat=np.matrix(L)
    mat.diagonal()
    B=np.zeros((n,n))
    for i in range(n):
        B[i,:]=mat[i][::-1]
    plt.imshow(B)

anti_diagonal([1,2,3,4])
```
